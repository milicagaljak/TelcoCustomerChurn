# Telco Curstomer Churn Prediction

`Problem koji se modelira u projektu je problem klasifikacije. Model pokušava da predvidi da li će korisnik telekomunikacionih usluga napustiti kompaniju ili ne. U projektu su utvrđeni i prikazani zaključci modela, koji atributi su najviše uticali na to da korisnik napusti kompaniju, a koji da ostane.`

`U projektu se demonstrira korišćenje biblioteke InterpretML. Biblioteka je specijalizovana za vizuelni prikaz odnosa promenljivih, uticaja na ciljnu promenljivu, raspodelu podataka, prikaz grešaka, prikaz ROC krive... Podržava ugrađene modele koje deli u dve grupe, na osnovu toga da li su modeli direktno interpretabilini ili ne. Grupa modela koja je direktno interpretabilna se naziva "Glassbox Models", dok je neinterpretabilna "Blackbox Explainers". U projektu su demonstrirani modeli iz grupe "Glassbox Models" kao što su Explainable Boosting Machine, Linear Model (LogisticRegression)...
Zvaničnoj stranici biblioteke se može pristupiti pomoću` [`linka`](https://interpret.ml/).

`Skup podataka je preuzet sa platforme Kaggle i može se pronaći na` [`adresi`](https://www.kaggle.com/datasets/blastchar/telco-customer-churn).

## Skup podataka

`Skup podataka koji se koristi u projektu je tabelarnog formata. Sadrži 20 kolona koji predstavljaju atribute i jednu kolonu koja predstavlja ciljnu promenljivu. Svaki red predstavlja podatke jednog korisnika. Skup podataka ima 7043 instanci.`

`Objašnjenje značajnijih atributa:`

`"tenure" - Predstavlja broj meseci koliko korisnik koristi telekomunikacione usluge.
"Contract" - Predstavlja tip ugovora koji je korisnik potpisao. Ugovor može biti dvogodišnji, jednogodišnji ili neobavezujući, koji se potpisuje iz meseca u mesec.
"MonthlyCharges" - Predstavlja količinu novca na mesečnom nivou koji se naplaćuje korisniku.
"TotalCharges" - Predstavlja ukupnu količinu novca koja je do sada naplaćena korisniku.
"Churn" - Predstavlja ciljnu promenljivu i označava da li je korisnik otišao iz kompanije ili nije.`

`Detaljan opis svakog atributa se može pronaći u projektu na putanji './dataset/DatasetDescription.docx'. `
